# Data Preprocessing

## Data Unpacking

```python
#unpack the loaction column
df_locs = df["location"].str.split(',',expand=True)
```

```python
#gets the country and district codes
df[["country_code", "district_code"]] = df_locs.iloc[:,:2]
```

```python
#there are sometimes dates added to the salary so they are removed
err_sal = ["Oct", "Jun", "Dec", "Apr", "Sep", "Nov"]
df.loc[df["salary_range"].str.split('-',expand=True)[0].isin(err_sal), "salary_range"]=None
df.loc[df["salary_range"].str.split('-',expand=True)[1].isin(err_sal), "salary_range"]=None
```

```python
#splits the data to minimum and maximum salary
df[["min_salary", "max_salary"]] = df["salary_range"].str.split('-',expand=True).astype(float)
```

## Data Quality: Missing values

```python
#prints the number of Missing values for each column
print("Missing data in columns")
print("--------------------------")
for col in df.columns[:22]:
    print(col+": "+ str(sum(df[col].isna())))
```

```python
#we can see the entry for description it has mostly missing data
df[df.description.isna()]
```

```python
#removed the entry with no description
df = df[~df.description.isna()]
```

```python
#for NA data for the country code and the district code changeit to not specified
df.loc[df.country_code.isna(), "country_code"]="Not Specified"
df.loc[df.district_code.isna(), "district_code"]="Not Specified"
```

```python
#for all the text based misssing values replace with a strign ""
text_cols = ['title', 'department', 'company_profile', 'description', 'requirements', 'benefits', 'industry', 'function']
df.loc[:,text_cols] = df.loc[:,text_cols].fillna("")
```

```python
#removing salary
df = df.drop(["salary_range", "min_salary", "max_salary"],axis =1)
```

## Data Quality: Outliers
As most of the data is either sting or boolean so there is no point for Outliers

## Data Quality: Duplicates

```python
df_dup = df[df.drop(["job_id"],axis =1).duplicated()]
print("Number of duplicated that are real job posts: ", str(sum(df_dup["fraudulent"]==0)))
print("Number of duplicated that are fraudalent job posts: ", str(sum(df_dup["fraudulent"]==1)))
```

```python
#removes duplicates
df = df[~df.drop(["job_id"],axis =1).duplicated()].reset_index(drop=True)
```

## Natural Language Processing
cleans the text data up

```python
df["text"] = df["title"]+" "+df["department"]+ " " + df["company_profile"]+ " " + df["description"]+ " "+ df["requirements"] +" "+ df["benefits"] + " "+ df["industry"]+" "+ df["function"]
```

```python
def nlp_text(text):
    #tokenizes the text to words
    words = nltk.tokenize.word_tokenize(text)
    
    #removes punctuation
    words = [w.lower() for w in words if w not in string.punctuation]
    
    #use lemmanizer for the words
    lemmer = WordNetLemmatizer()
    lemm_words = [lemmer.lemmatize(word) for word in words]

    #remove the stopwords
    stopwords = nltk.corpus.stopwords.words("english")
    lemm_stop = [word for word in lemm_words if word not in stopwords]
    
    return(" ".join(lemm_stop))
```

```python
df["text"] = df["text"].apply(nlp_text)
```

```python
voc_size=5000
onehot_repr=[one_hot(words,voc_size) for words in df["text"]] 
```

```python
sent_length=50
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen = sent_length)
```

```python
X_words=np.array(embedded_docs)
```

## Multi Layered Perceptron

```python
def metrics(y_true, y_pred, name):
    print("Accuracy: "+ str(accuracy_score(y_true, y_pred)))
    print("Recall: "+ str(recall_score(y_true, y_pred)))
    print("Precision: "+ str(precision_score(y_true, y_pred)))
    print("F1_Score: "+ str(f1_score(y_true, y_pred)))
    
    
    cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_true, y_pred), display_labels = ["Real", "Fraudalent"])

    cm_display.plot()
    plt.title("Confusion Matrix of "+ name)
    plt.show()
    
    return(pd.DataFrame([accuracy_score(y_true, y_pred), recall_score(y_true, y_pred), precision_score(y_true, y_pred),f1_score(y_true, y_pred)], columns = [name]))
          
```

```python
model1 = Sequential()
model1.add(Dense(512, input_dim=X_train.shape[1], activation='relu'))
model1.add(Dense(900, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer=optimizers.Adamax(),
              metrics=['accuracy', Recall(), Precision(), tf.keras.metrics.AUC()]) 
print(model1.summary())
```

```python
#using the word data
model1.fit(np.asarray(X_train),np.asarray(y_train),epochs=24,batch_size=64)
```

```python
y_pred_mlp=(model1.predict(np.asarray(X_test)) > 0.5).astype("int32")
```

```python
df_mlp = metrics(y_test, y_pred_mlp, "MLP")
```
